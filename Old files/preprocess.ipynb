```python
import pandas as pd

# Load the CSV file
df = pd.read_csv('cleanposts.csv')
```

```python
df.head(5)
```

```python
df_cleaned = df.drop_duplicates()

df_cleaned = df.dropna()

# Drop unnecessary columns
#df_cleaned = df.drop(['threadid', 'authorid', 'postid', 'forumid'], axis=1)
```

```python
df_cleaned.head(10)
```

```python
# Step 1: Display rows where the 'lang' column is not 'en' or 'nl'
mask = ~df_cleaned['lang'].isin(['en', 'nl'])
print(df_cleaned[mask].head(5))
```

```python
# Step 2: Remove rows where the 'lang' column is not 'en'
df_cleaned = df_cleaned[df_cleaned['lang'] == 'en']
```

```python
# check if removed
mask = ~df_cleaned['lang'].isin(['en', 'nl'])
print(df_cleaned[mask].head(5))
```

```python
with open('interpersonal_relationships_keywords.txt', 'r') as file:  # Update this path
    keywords = [line.strip() for line in file.readlines()]

print(keywords)
```

```python
# Define a function to search for any keyword within a message, case-sensitive
def contains_any_keyword(message, keywords):
    return any(keyword in message for keyword in keywords)

# Apply the function to the 'cleanmessage' column to create a mask
mask = df_cleaned['cleanmessage'].apply(lambda message: contains_any_keyword(message, keywords))

# Use the mask to filter the DataFrame
filtered_df = df_cleaned[mask]

# Optionally, inspect the first few rows of the filtered DataFrame
print(filtered_df.head())
```
